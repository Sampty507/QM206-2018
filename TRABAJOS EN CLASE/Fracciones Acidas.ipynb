```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fraccion(pH,pKa):
    H=10**(-pH)
    Ka=[1]
    for pk in pKa:
        x=10**(-pk)
        Ka.append(x)
    n=len(Ka)-1
    mult=1
    terminos=[]
    for a in range(n+1):
        n1=(n-a)
        cH=H**n1
        mult=mult*Ka[a]
        d=cH*mult
        terminos.append(d)
    sumaTerminos=sum(terminos)
    print(terminos/sumaTerminos) #ya da la fracción
    alpha=terminos/sumaTerminos #función de la fracción
    return alpha # ya genera la fracción para cada pH
```

```python
pKa=np.array([4.0,6.7,8.5])
pH=0
x=[]
y=[]
y0=[]
y1=[]
y2=[]
y3=[]
while pH<=14: #ciclo para calcular el pH en un aumento de 0.1 en 0.1
    x.append(pH)
    alpha=fraccion(pH,pKa)
    y.append(alpha[0])  
    y0.append(alpha[0])
    y1.append(alpha[1])
    y2.append(alpha[2])
    y3.append(alpha[3])
    plt.plot(pH,alpha[0])
    pH+=0.1 #recordar poner esta linea sino se tildea la PC lollll
plt.plot(x,y0)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.xlabel("pH")
plt.ylabel("Fracciones de las especies")
plt.suptitle("Fracciones de especies ácidas")
plt.show()

```
