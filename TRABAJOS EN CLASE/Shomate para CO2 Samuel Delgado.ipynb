```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
T=np.arange(100,850,50)
Cp=[40.144,41.826,43.396,44.860,46.222,47.489,48.665,49.755,50.764,51.698,52.559,53.353,54.086,54.755,55.371]
```

```python
T=T+273.15 #transformar de grados centigrados a kelvin
y=Cp
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
print(beta)
Cp_calc=X.dot(beta)
Cp_calc #Cp calculados
```

```python
plt.scatter(T,y,color="red")
plt.plot(T,Cp_calc)
plt.show()
```

```python
dT=0.1
T=np.arange(100,800,dT)
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
Cp_calc=X.dot(beta)
plt.plot(T,Cp_calc)
plt.show()
```

```python
H=np.cumsum(Cp_calc*dT)
plt.plot(T,H)
plt.xlabel("Temperatura(K)")
plt.ylabel("delta_H (J/mol)")
plt.title("T vs H")
plt.show()
```

```python
S=np.cumsum(Cp_calc/T*dT)
G=H-T*S
plt.plot(T,G)
plt.xlabel("T(K)")
plt.ylabel("delta_G(J/mol)")
plt.title("delta_G vs T")
plt.show()
```

```python
plt.plot(T,S)
plt.xlabel("T(K)")
plt.ylabel("delta_S (J/K-mol)")
plt.title("delta_S vs T")
plt.show()
```
