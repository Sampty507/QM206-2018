```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
```

```python
iris = datasets.load_iris()
iris_dataframe=pd.DataFrame(data=iris.data,columns=["SL","SW","PL","PW"])
iris_dataframe["target"]=iris.target#target es un arreglo de 150
x=iris_dataframe["SL"]
print(x.mean())
y=iris_dataframe["SW"]
col=(iris_dataframe["target"]+1)*2
print(y.mean())
PL=iris_dataframe["PL"]*20
PW=iris_dataframe["PW"]*60
```

```python
plt.scatter(x,y,c=col,s=PL,alpha=(0.5))
plt.xlabel("Largo del Sépalo(cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.suptitle("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```

```python
plt.scatter(x,y,c=col,s=PW,alpha=(0.70))
plt.xlabel("Largo del Sépalo(cm)")
plt.ylabel("Ancho del Sépalo(cm)")
plt.suptitle("Ancho del Sépalo vs Largo del Sépalo")
plt.grid(True)
plt.show()
```
