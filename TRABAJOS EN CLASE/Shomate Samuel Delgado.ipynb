```python
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
#coeficiente de shomate para el diazeno n2h2
A=8.8997
B=93.7976
C=-53.12099
D=12.1159
E=0.36272
H298=55.13 #KJ/mol
```

```python
dT=0.1
T=np.arange(298,1400,0.1)
t=T/1000

Cp=A+B*t+C*t**2+D*t**3+E/t**2 #unidades J/K-mol
H=np.cumsum(Cp*dT)+55.13*1000
S=np.cumsum(Cp/T*dT)
G=H-T*S
```

```python
plt.plot(T,Cp,color="green")
plt.title("Cp vs T")
plt.xlabel("Temperatura (K)")
plt.ylabel("Cp ( J / K-mol )")
plt.show()
```

```python
plt.plot(T,H,color="blue") #aumento de calor
plt.plot(T,S,color="green") #aumento de desorden
plt.plot(T,G,color="purple") 
plt.title(" entropía, entalpía, energía libre")
plt.ylabel("H,G,S")
plt.xlabel("Temperatura (K)")
plt.show() #esta grafica nos indica que el delta H aumenta aumentar calor y la energia libre tambien aumenta entre mas negativa mayor trabajo
```
