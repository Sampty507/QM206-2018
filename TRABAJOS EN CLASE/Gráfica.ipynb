```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([0,2,4,1],[3,1,7,4],color="deepskyblue",linewidth=2,alpha=0.5,marker="o",markersize=5,markeredgecolor="darkmagenta",markerfacecolor="blue",markeredgewidth=3)
plt.xlabel("$x(cm^2)$")
plt.ylabel("$Intensidad(\mu)$")
plt.title("Experimento No. 1")
plt.grid(True)
plt.show()
```

```python
x=np.arange(0,2*np.pi,0.1)
y=np.sin(4*x)
y2=np.cos(2*x)
plt.plot(x,y, color="tomato")
plt.plot(x,y2,color="dodgerblue", dashes=[6,3,6,3])
plt.title("Funcin sen (4x) 0-$2\pi$")
plt.grid(True)
plt.xlabel("t (seg)")
plt.ylabel("Amplitud de la onda")
plt.show()
```

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
plt.scatter(x,y,s=r,color="red")
r=np.random.rand(100)*250+50
plt.show
```

```python
x=np.random.randn(10000)
plt.hist(x,bins=50,facecolor="blueviolet",edgecolor="cyan",density=True)
plt.title("Distribucin de la poblacin")
plt.ylabel("Densidad habitantes/$m^2$")
plt.text(-3, .35, r'$\mu=100,\ \sigma=15$')
plt.show()
```
