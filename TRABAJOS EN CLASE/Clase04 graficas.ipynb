```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(10)
y=np.sin(x)
plt.plot(x,y,"ro-")
plt.xlabel("distancia(m)")
plt.ylabel("densidad (g/ml)")
plt.suptitle("Experimento 1")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot([1,4,2,6],"ro-.")
plt.subplot(2,2,2)
plt.hist(np.random.randn(1000),color="orange")
plt.subplot(2,2,3)
plt.plot(np.random.rand(10))
plt.xlabel("x3(cm)")
plt.subplot(2,2,4)
plt.plot([1,2,3])
plt.suptitle("Datos Experimentales")
plt.show()
```
