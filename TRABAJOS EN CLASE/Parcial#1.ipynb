```python

A=3.52
B=-2.85
C=5.16
x=np.arange(0, 10)
y=A+10**(x)*B+C*x**2/2
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y vs x")
plt.show()


    
```

```python
Ca=1.0
Cb=0.01
pKa=4.72
x=np.arange(0.1,0.9,0.01)
pH=pKa+10**((Ca-x)/(Cb+x))
y=pH
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("pH")
plt.title("pH vs x")
plt.show()
```
