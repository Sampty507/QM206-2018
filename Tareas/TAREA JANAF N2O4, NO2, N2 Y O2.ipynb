```python
import numpy as np
import matplotlib.pyplot as plt
import requests
```

## N2O4

```python
def LeerArchivo(url):
#url="https://janaf.nist.gov/tables/N-032.txt" #como llamarlo dsde un url
    file=requests.get(url)
    datos=file.text
    datos=datos.split("\n")[3:-1]
    mat=[]
    for x in datos:
        s=x.split("\t")
        mat.append([float(a)for a in s])
    mat=np.array(mat)
    return mat
```

```python
mat=LeerArchivo("https://janaf.nist.gov/tables/N-032.txt")
T=mat[:,0]
Cp=mat[:,1]
plt.plot(T,Cp,color="yellow")
plt.show()
```

```python
T=T[3:16]
y=Cp[3:16]
t=T/1000
n=len(T)
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
beta
```

```python
Cp_calc=beta.dot(X.T)
```

```python
plt.scatter(T,y)
plt.plot(T,Cp_calc,color="yellow")
plt.xlabel("T(K)")
plt.ylabel("Cp(J/K-mol)")
plt.show()
```

## N2

```python
def LeerArchivo(url):
    file=requests.get(url)
    datos=file.text
    datos=datos.split("\n")[3:-1]
    mat=[]
    for x in datos:
        s=x.split("\t")
        mat.append([float(a)for a in s])
    mat=np.array(mat)
    return mat
```

```python
mat=LeerArchivo("https://janaf.nist.gov/tables/N-023.txt")
T=mat[:,0]
Cp=mat[:,1]
plt.plot(T,Cp,COLOR="orange")
plt.show()
```

```python
T=T[3:16]
y=Cp[3:16]
t=T/1000
n=len(T)#cuantas filas hay en los datos
X=np.array([np.ones(n),t,t**2,t**3,1/t**2]).T
beta=np.linalg.inv(X.T.dot(X)).dot(X.T.dot(y))
Cp_calc=X.dot(beta)
plt.scatter(T,y)
plt.plot(T,Cp_calc,color="orange")
plt.xlabel("T(K)")
plt.ylabel("Cp(J/K-mol)")
plt.title("Cp vs T")
plt.show()
```
