```python
wine = datasets.load_wine()
wine_dataframe=pd.DataFrame(data=wine.data,columns=wine.feature_names)
wine_dataframe
```

### Base de datos Wine
Estos datos son el resultado de un análisis químico de vinos cultivados en la misma región en Italia pero derivados de tres cultivares diferentes. El análisis determinó las cantidades de 13 componentes encontrados en cada uno de los tres tipos de vinos.

Los atributos son (donados por Riccardo Leardi, riclea '@' anchem.unige.it)

Alcohol
Ácido málico
Ceniza
Alcalinidad de ceniza
Magnesio
Fenoles totales
Flavonoides
Fenoles no flavonoides
Proanthocyanins 10) Color intensidad 11) Hue 12) OD280 / OD315 de vinos diluidos 13) Proline
En un contexto de clasificación, este es un problema bien planteado con estructuras de clase "de buen comportamiento". Un buen conjunto de datos para la primera prueba de un nuevo clasificador, pero no muy desafiante.

```python
wine_dataframe["target"]=wine.target
x=wine_dataframe["alcohol"]
print(x.mean())
y=wine_dataframe["malic_acid"]
col=(wine_dataframe["target"]+1)*2
print(y.mean())
alcohol=wine_dataframe["alcohol"]*30
malic_acid=wine_dataframe["malic_acid"]*30
print(x.std())
print(y.std())
```

```python
plt.scatter(x,y,c=col,s=200,alpha=(0.6))
plt.xlabel("Alcohol(mL)")
plt.ylabel("Ácido málico(mL)")
plt.suptitle("Ácido málico vs Alcohol")
plt.grid(True)
plt.show()
```

### Conclusión
En esta gráfica el vino(amarillo) se aprecia como se da el aumento de concentración de ácido málico a medida que aumenta el alcohol.
